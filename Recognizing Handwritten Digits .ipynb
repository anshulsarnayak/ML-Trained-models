```python
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
dig=load_digits()
dir(dig)
```

```python
dig.images[0]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
for i in range(10):
    plt.matshow(dig.images[i])
```

```python
plt.imshow(dig.images[0], cmap=plt.cm.gray_r, interpolation='nearest')
```

```python
svc = svm.SVC(gamma=0.001, C=100.)
svc.fit( dig.data[:1500] ,dig.target[:1500])
```

```python
plt.imshow(dig.images[999], cmap=plt.cm.gray_r,
interpolation='nearest')
```

```python
svc.predict(dig.data[999:1000])
```

```python
y_pred = svc.predict(dig.data[1500:])
y_test = dig.target[1500:]
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)

%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```
